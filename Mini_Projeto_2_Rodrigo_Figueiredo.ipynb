```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Walmart.csv')

df
```

```python
df.describe()
```

```python
store = df.groupby(['Store']).mean().reset_index()

store.sort_values(by='Weekly_Sales', ascending=False)
```

# A partir do mÃ©todo de correlaÃ§Ã£o nota-se que o preÃ§o do combustivel Ã© a metrica que mais se correlaciona com as vendas semanais, entretanto, nÃ£o parece ser uma medida relevante para a anÃ¡lise em questÃ£o.

```python
corr = store.corr()

f, ax = plt.subplots(figsize=(11, 9))
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
